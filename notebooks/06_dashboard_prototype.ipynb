# Dashboard Prototype

Prototype dashboard components inline before deploying to Streamlit.

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go

from dashboard.components.charts import heatmap, line_chart, bar_chart

rng = np.random.default_rng(42)
assets = ['SPX', 'NDX', 'HY_OAS', 'VIX', 'COPPER', 'DXY', 'BTC']
n = len(assets)

# Synthetic TE matrix
te_data = np.abs(rng.random((n, n))) * 0.15
np.fill_diagonal(te_data, 0.0)
te_matrix = pd.DataFrame(te_data, index=assets, columns=assets)

fig = heatmap(te_matrix, title='Transfer Entropy Matrix (lag=1)', zmid=0)
fig.show()
```

```python
# Regime probability chart
T = 300
dates = pd.date_range('2021-01-01', periods=T, freq='B')
prob_regime0 = pd.Series(rng.dirichlet([3, 1], T)[:, 0], index=dates)
prob_regime1 = 1 - prob_regime0

fig = line_chart(
    {'Regime 0 (Low Vol)': prob_regime0, 'Regime 1 (High Vol)': prob_regime1},
    title='Regime Probabilities',
    y_label='Probability',
)
fig.show()
```

```python
# Signal attribution bar chart
signals = {
    'HY_OAS→SPX': -0.003,
    'COPPER→XLI': 0.002,
    'VIX→HY_OAS': -0.001,
    'BTC→ETH': 0.005,
}
fig = bar_chart(signals, title='Signal Expected Returns', y_label='Return')
fig.show()
```
