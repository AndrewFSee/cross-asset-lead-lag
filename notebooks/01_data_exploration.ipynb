# Data Exploration

Load data, check coverage, visualize correlations, and check stationarity.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from data.preprocessing import stationarity_check, winsorize_returns
from data.returns import build_returns_panel

%matplotlib inline
print('Imports OK')
```

## Generate Synthetic Data (no API key needed)

```python
rng = np.random.default_rng(42)
T = 504
dates = pd.date_range('2020-01-01', periods=T, freq='B')

assets = ['SPX', 'NDX', 'HY_OAS', 'VIX', 'COPPER', 'DXY', 'BTC', 'UST_10Y']
returns = pd.DataFrame(
    rng.standard_normal((T, len(assets))) * 0.01,
    index=dates, columns=assets
)
print(f'Returns panel: {returns.shape}')
returns.head()
```

## Stationarity Check

```python
stat_results = {}
for col in returns.columns:
    stat_results[col] = stationarity_check(returns[col])

stat_df = pd.DataFrame(stat_results).T
stat_df[['adf_pvalue', 'kpss_pvalue', 'is_stationary']]
```

## Correlation Heatmap

```python
fig, ax = plt.subplots(figsize=(10, 8))
corr = returns.corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='RdBu_r', center=0, ax=ax)
ax.set_title('Return Correlations')
plt.tight_layout()
plt.show()
```
