# Transfer Entropy Analysis

Compute TE matrix and identify strongest lead-lag relationships.

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

from discovery.transfer_entropy import compute_te_matrix, transfer_entropy_knn

rng = np.random.default_rng(42)
T = 300

# Asset A leads Asset B (causal)
A = rng.standard_normal(T)
B = np.zeros(T)
for t in range(1, T):
    B[t] = 0.6 * A[t-1] + 0.4 * rng.standard_normal()

C = rng.standard_normal(T)  # Independent

print('Synthetic data generated')
```

```python
# Compute TE
te_ab = transfer_entropy_knn(A, B, lag=1, k=5)
te_ba = transfer_entropy_knn(B, A, lag=1, k=5)
print(f'TE(A→B) = {te_ab:.4f}  (expected: high, A leads B)')
print(f'TE(B→A) = {te_ba:.4f}  (expected: low)')
```

```python
# Compute full TE matrix
returns_dict = {'A': A, 'B': B, 'C': C}
te_matrices = compute_te_matrix(returns_dict, lags=[1, 2, 5], k=5)

fig, axes = plt.subplots(1, 3, figsize=(14, 4))
for ax, (lag, df) in zip(axes, te_matrices.items()):
    sns.heatmap(df, annot=True, fmt='.3f', cmap='YlOrRd', ax=ax, vmin=0)
    ax.set_title(f'TE Matrix (lag={lag})')
plt.tight_layout()
plt.show()
```
