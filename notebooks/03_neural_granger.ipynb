# Neural Granger Causality

Train ComponentLSTM and measure ablation-based GC scores.

```python
import numpy as np
import matplotlib.pyplot as plt

from discovery.neural_granger import ComponentLSTM, neural_granger_test

rng = np.random.default_rng(42)
T = 400

A = rng.standard_normal(T)
B = np.zeros(T)
for t in range(1, T):
    B[t] = 0.7 * A[t-1] + 0.3 * rng.standard_normal()
C = rng.standard_normal(T)

data = np.column_stack([A, B, C])  # B is target (index 1)
print(f'Data shape: {data.shape}')
```

```python
# Run neural Granger test targeting B (index 1)
gc_scores = neural_granger_test(
    data,
    target_idx=1,
    n_epochs=50,
    history_len=10,
    hidden_dim=32,
    embed_dim=8
)

print('Neural GC Scores (target = B):')
labels = {0: 'A (true cause)', 1: 'B (self)', 2: 'C (noise)'}
for idx, score in gc_scores.items():
    print(f'  {labels.get(idx, str(idx))}: {score:.4f}')
```

```python
# Visualize GC scores
fig, ax = plt.subplots(figsize=(7, 4))
items = [(labels.get(k, str(k)), v) for k, v in gc_scores.items()]
ax.bar([x[0] for x in items], [x[1] for x in items], color=['green', 'blue', 'red'])
ax.set_title('Neural GC Scores → Target B')
ax.set_ylabel('GC Score (relative MSE increase)')
plt.tight_layout()
plt.show()
```
